# Load case9 with MATLAB Engine

This notebook loads `case9.m` using the **MATLAB** engine and runs power flow via `to_mpc(backend="matlab")`.

> **Requirement:** MATLAB Engine for Python must be installed (`import matlab.engine`).
> The cell below checks availability and skips if not found.

```python
%load_ext autoreload
%autoreload 2
```

```python
from matpower import run_matlab_cmd, start_instance

from matpowercaseframes import CaseFrames
```

```python
import pytest

try:
    import matlab.engine  # noqa: F401

    MATLAB_AVAILABLE = True
except ImportError:
    MATLAB_AVAILABLE = False

if not MATLAB_AVAILABLE:
    raise pytest.skip("MATLAB not available")
```

```python
m = start_instance(engine="matlab")
```

```python
CASE_NAME = "case9.m"
cf_9 = CaseFrames(CASE_NAME, load_case_engine=m)
cf_9.infer_numpy()
cf_9.branch
```

```python
# NOTE: need "matlab" backend due to matlab didn't support plain python types
mpc = cf_9.to_mpc(backend="matlab")
mpc
```

```python
# Run power flow using the converted mpc from CaseFrames
r1 = run_matlab_cmd("runpf(mpc)", m=m, mpc=mpc)
r1
```

```python
m.exit()
```
