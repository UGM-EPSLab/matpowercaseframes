# Load case_RTS_GMLC

This notebook loads the RTS-GMLC case which contains:
- Piecewise linear gencost (`TYPE=1`)
- `bus_name` field

This notebook loads `RTS-GMLC.m` using the **MATLAB** engine and runs power flow via `to_mpc(backend="matlab")`.

> **Requirement:** MATLAB Engine for Python must be installed (`import matlab.engine`).
> The cell below checks availability and skips if not found.

```python
%load_ext autoreload
%autoreload 2
```

```python
from matpower import run_matlab_cmd, start_instance

from matpowercaseframes import CaseFrames
```

```python
import pytest

try:
    import matlab.engine  # noqa: F401

    MATLAB_AVAILABLE = True
except ImportError:
    MATLAB_AVAILABLE = False

if not MATLAB_AVAILABLE:
    raise pytest.skip("MATLAB not available")
```

```python
m = start_instance(engine="matlab")
```

```python
CASE_NAME = "case_RTS_GMLC.m"
cf_RTS_GMLC = CaseFrames(CASE_NAME, load_case_engine=m)
display(cf_RTS_GMLC.gencost.head())  # case_RTS_GMLC use linear cost function
display(cf_RTS_GMLC.bus.head())  # case_RTS_GMLC use bus_name
```

```python
# NOTE: bus_name is not supported in MATLAB
# See: https://github.com/mathworks/matlab-engine-for-python/issues/61
mpc = cf_RTS_GMLC.to_mpc(backend="matlab")
mpc
```

```python
r1 = run_matlab_cmd("runpf(mpc)", m=m, mpc=mpc)
r1
```

```python
m.exit()
```
